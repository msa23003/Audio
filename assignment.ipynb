### TASK 1: Text to Speech


```python
import re
from google.cloud import texttospeech
from langdetect import detect, DetectorFactory
from google.api_core.exceptions import InvalidArgument

# Ensure consistent results for language detection
DetectorFactory.seed = 0

LANGUAGE_CODE_MAP = {
    "en": "en-US",
    "hi": "hi-IN",
    "bn": "bn-IN",
    "mr": "mr-IN",
    "te": "te-IN",
    "gu": "gu-IN",
    "kn": "kn-IN",
    "ta": "ta-IN",
    "ur": "ur-IN",
}

def detect_language(text: str) -> str:
    """
    Detect the primary language of the input text.
    """
    return detect(text)

def normalize_text(html_text: str) -> str:
    """
    Normalize HTML content for better TTS processing.
    """
    html_text = re.sub(r"&", "and", html_text)
    html_text = re.sub(r"\s+", " ", html_text)
    html_text = re.sub(r"([A-Za-z])\.([A-Za-z])", r"\1. \2", html_text)
    return html_text

def html_to_ssml(input_html: str, method="rule-based") -> str:
    """
    Convert HTML content to SSML format.
    """
    html_content = normalize_text(input_html)
    ssml = "<speak>"

    html_content = re.sub(r"<h[1-6][^>]*>(.*?)</h[1-6]>",
                          r"<p><emphasis level=\"moderate\">\1</emphasis></p>", html_content)
    html_content = re.sub(r"<p[^>]*>(.*?)</p>", r"<p>\1</p>", html_content)
    html_content = re.sub(r"<b>(.*?)</b>", r"<emphasis level=\"strong\">\1</emphasis>", html_content)
    html_content = re.sub(r"<li>(.*?)</li>", r"<s>\1</s><break time=\"200ms\"/>", html_content)
    html_content = re.sub(r"</p>\s*<p>", r"</p><break time=\"300ms\"/><p>", html_content)
    html_content = re.sub(r"<[^>]+>", "", html_content)

    ssml += html_content + "</speak>"
    return ssml

def get_available_voices(language_code: str):
    """
    Retrieve and filter voices for the specified language code.
    """
    client = texttospeech.TextToSpeechClient()
    response = client.list_voices()
    return [
        voice for voice in response.voices
        if language_code in voice.language_codes
        and voice.ssml_gender in (texttospeech.SsmlVoiceGender.MALE, texttospeech.SsmlVoiceGender.FEMALE)
    ]

def text_to_speech_with_ssml(ssml_text: str, output_file: str, voice: str, pitch=0.0, rate=1.0, volume=0.0):
    """
    Convert SSML text to speech and save the output as an audio file.
    """
    client = texttospeech.TextToSpeechClient()
    lang_code = "-".join(voice.split("-")[:2])

    voice_config = texttospeech.VoiceSelectionParams(
        language_code=lang_code,
        name=voice,
    )
    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.MP3,
        speaking_rate=rate,
        pitch=pitch,
        volume_gain_db=volume
    )

    try:
        synthesis_input = texttospeech.SynthesisInput(ssml=ssml_text)
        response = client.synthesize_speech(
            input=synthesis_input,
            voice=voice_config,
            audio_config=audio_config
        )
        with open(output_file, "wb") as out:
            out.write(response.audio_content)
        print(f"Audio content written to '{output_file}'")
    except InvalidArgument as e:
        print(f"Error generating audio: {e}")

def main():
    print("Welcome to the HTML-to-Speech Generator")

    input_html = input("Enter your HTML content: ")
    detected_lang = detect_language(input_html)
    print(f"Detected Language: {detected_lang}")

    google_tts_lang_code = LANGUAGE_CODE_MAP.get(detected_lang, "en-US")
    voices = get_available_voices(google_tts_lang_code)

    if not voices:
        print(f"No voices available for '{google_tts_lang_code}'. Exiting.")
        return

    print("\nAvailable Voices:")
    for i, voice in enumerate(voices):
        print(f"{i + 1}. {voice.name} - {voice.language_codes}, Gender: {voice.ssml_gender}")

    try:
        selected_index = int(input("\nSelect a voice by number (default=1): ").strip() or 1)
        if not (1 <= selected_index <= len(voices)):
            raise ValueError
    except ValueError:
        print("Invalid selection. Defaulting to 1.")
        selected_index = 1

    chosen_voice = voices[selected_index - 1].name
    print(f"\nChosen voice: {chosen_voice}")

    method = input("Choose conversion method (rule-based/llm) [default: rule-based]: ").strip().lower() or "rule-based"
    pitch = float(input("Enter pitch (-20 to 20, default=0): ").strip() or 0.0)
    rate = float(input("Enter rate (0.5 to 2.0, default=1.0): ").strip() or 1.0)
    volume = float(input("Enter volume (-10 to 10, default=0.0): ").strip() or 0.0)

    ssml_text = html_to_ssml(input_html, method=method)
    output_file = "output.mp3"
    print(f"\nGenerating audio using {chosen_voice} with pitch={pitch}, rate={rate}, volume={volume}")
    text_to_speech_with_ssml(ssml_text, output_file, chosen_voice, pitch, rate, volume)


if __name__ == "__main__":
    main()
```

```python
import os
import re
from yt_dlp import YoutubeDL
from pydub import AudioSegment
from google.cloud import speech
from google.cloud import texttospeech

import librosa
import numpy as np
import soundfile as sf
import google.generativeai as genai

# API Key Configuration
API_KEY = "AIzaSyBUXqqE8lAkEdsJFO6dhEn12CUZgYCO3dY"
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/verdant-upgrade-448910-r0-a846d4f1e5c8.json"  # Update this
genai.configure(api_key=API_KEY)

# Predefined list of target languages
TARGET_LANGUAGES = {
    "1": ("Hindi", "hi"),
    "2": ("Bengali (Bangla)", "bn"),
    "3": ("Marathi", "mr"),
    "4": ("Odia", "or"),
    "5": ("Telugu", "te"),
    "6": ("Arabic (Modern)", "ar"),
    "7": ("Urdu", "ur"),
    "8": ("Persian", "fa"),
    "9": ("Spanish", "es"),
    "10": ("Portuguese", "pt")
}

LANGUAGE_CODE_MAP = {lang[1]: f"{lang[1]}-IN" for lang in TARGET_LANGUAGES.values()}

def download_youtube_audio(youtube_url):
    ydl_opts = {
        "format": "bestaudio/best",
        "postprocessors": [{
            "key": "FFmpegExtractAudio",
            "preferredcodec": "wav",
            "preferredquality": "192",
        }],
        "outtmpl": "youtube_audio.%(ext)s",
    }

    with YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])

    audio_path = "youtube_audio.wav"
    audio = AudioSegment.from_file(audio_path)
    audio = audio.set_frame_rate(16000)  # Ensure 16,000 Hz sample rate
    audio.export("audio.wav", format="wav")
    return "audio.wav"

def split_audio(input_audio_path, chunk_duration_ms=10000):
    audio = AudioSegment.from_file(input_audio_path)
    chunks = [audio[i:i + chunk_duration_ms] for i in range(0, len(audio), chunk_duration_ms)]

    chunk_paths = []
    for idx, chunk in enumerate(chunks):
        chunk_path = f"chunk_{idx + 1}.wav"
        chunk.export(chunk_path, format="wav")
        chunk_paths.append(chunk_path)

    return chunk_paths

def transcribe_audio(audio_path):
    client = speech.SpeechClient()
    y, sr = librosa.load(audio_path, sr=16000)
    resampled_audio_path = "resampled_audio.wav"
    sf.write(resampled_audio_path, y, sr)

    with open(resampled_audio_path, "rb") as audio_file:
        content = audio_file.read()

    audio = speech.RecognitionAudio(content=content)
    config = speech.RecognitionConfig(
        encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
        sample_rate_hertz=16000,
        language_code="en-US",
    )

    response = client.recognize(config=config, audio=audio)
    transcription = " ".join([result.alternatives[0].transcript for result in response.results])
    return transcription

def translate_html(html_content, target_language):
    system_instruction = f"""
    You are a highly capable translation model.
    Translate the text inside the HTML tags to {target_language}.
    Keep all HTML tags and formatting intact, but translate the content within the tags.
    """
    user_content = f"HTML content to translate:\n{html_content}"
    prompt_text = system_instruction.strip() + "\n\n" + user_content.strip()

    print("\nTranslating the content... This may take a few seconds.")
    model = genai.GenerativeModel(model_name="gemini-1.5-flash")
    response = model.generate_content(prompt_text)

    return response.text

def html_to_ssml(input_html):
    from bs4 import BeautifulSoup

    # Parse the HTML to extract the inner text
    soup = BeautifulSoup(input_html, "html.parser")
    text_content = soup.get_text()  # Extract the text content without HTML tags

    # Do not escape the actual text content; wrap only in <speak> tags
    ssml = f"<speak>{text_content.strip()}</speak>"
    return ssml




def split_ssml(ssml_text, max_bytes=5000):
    chunks = []
    current_chunk = ""
    current_chunk_size = 0

    # Remove extra <speak> tags in the input text if present
    ssml_text = re.sub(r"<speak>|</speak>", "", ssml_text)

    # Split text into sentences using punctuation
    sentences = re.split(r'(?<=[.!?])\s+', ssml_text)

    for sentence in sentences:
        sentence = sentence.strip()
        sentence_size = len(sentence.encode("utf-8"))

        if sentence_size > max_bytes - len("<speak></speak>"):
            raise ValueError(f"A single sentence exceeds the {max_bytes}-byte limit. Please shorten the input.")

        if current_chunk_size + sentence_size + len("<speak></speak>") > max_bytes:
            # Finalize the current chunk
            chunks.append(f"<speak>{current_chunk.strip()}</speak>")
            current_chunk = ""
            current_chunk_size = 0

        current_chunk += sentence + " "
        current_chunk_size += sentence_size

    # Add the last chunk if not empty
    if current_chunk.strip():
        chunks.append(f"<speak>{current_chunk.strip()}</speak>")

    return chunks



def get_available_voices(language_code):
    client = texttospeech.TextToSpeechClient()
    response = client.list_voices()
    return [voice for voice in response.voices if language_code in voice.language_codes]

from pydub import AudioSegment

def text_to_speech_with_ssml(ssml_text, output_file, voice, pitch, rate, volume, retries=3):
    client = texttospeech.TextToSpeechClient()
    lang_code = "-".join(voice.split("-")[:2])

    voice_config = texttospeech.VoiceSelectionParams(
        language_code=lang_code,
        name=voice,
        ssml_gender=texttospeech.SsmlVoiceGender.FEMALE
    )
    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.MP3,
        speaking_rate=rate,
        pitch=pitch,
        volume_gain_db=volume
    )

    # Split SSML into manageable chunks
    ssml_chunks = split_ssml(ssml_text)

    # Generate audio for each SSML chunk
    temp_file = "temp_output.mp3"
    with open(temp_file, "wb") as out:
        for ssml_chunk in ssml_chunks:
            success = False
            for attempt in range(retries):
                try:
                    synthesis_input = texttospeech.SynthesisInput(ssml=ssml_chunk)
                    response = client.synthesize_speech(
                        input=synthesis_input,
                        voice=voice_config,
                        audio_config=audio_config
                    )
                    out.write(response.audio_content)
                    success = True
                    break
                except Exception as e:
                    print(f"Error during TTS (attempt {attempt + 1}/{retries}): {e}")

            if not success:
                raise RuntimeError(f"Failed to generate audio for SSML chunk after {retries} retries.")

    # Trim the first 2.5 seconds and the last 2 seconds
    audio = AudioSegment.from_file(temp_file)
    start_trim = 2500  # Skip the first 2.5 seconds (2500 milliseconds)
    end_trim = 2000  # Skip the last 2 seconds (2000 milliseconds)
    trimmed_audio = audio[start_trim:len(audio) - end_trim]

    # Export the final trimmed audio
    trimmed_audio.export(output_file, format="mp3")


if __name__ == "__main__":
    youtube_url = input("Enter YouTube video URL: ")
    target_lang_choice = input("\nSelect a target language for translation:\n" + "\n".join([f"{k}. {v[0]}" for k, v in TARGET_LANGUAGES.items()]) + "\n\nEnter the number corresponding to your choice: ")

    target_language = TARGET_LANGUAGES.get(target_lang_choice, ("Hindi", "hi"))[0]
    audio_path = download_youtube_audio(youtube_url)

    chunks = split_audio(audio_path)

    transcription = ""
    for chunk in chunks:
        transcription += transcribe_audio(chunk) + " "

    translated_text = translate_html(transcription, target_language)
    ssml_text = html_to_ssml(translated_text)

    available_voices = get_available_voices(LANGUAGE_CODE_MAP.get(target_lang_choice, "hi-IN"))
    print("\nAvailable voices:")
    for idx, voice in enumerate(available_voices, 1):
        print(f"{idx}. {voice.name}")

    voice_choice = int(input("Select a voice by number (default=1): ") or 1)
    selected_voice = available_voices[voice_choice - 1].name

    pitch = float(input("Enter pitch (-20 to 20, default=0): ") or 0)
    rate = float(input("Enter rate (0.5 to 2.0, default=1.0): ") or 1.0)
    volume = float(input("Enter volume gain in dB (-16 to 16, default=0): ") or 0)

    output_file = "output_audio.mp3"
    text_to_speech_with_ssml(ssml_text, output_file, selected_voice, pitch, rate, volume)

    print(f"Audio file generated: {output_file}")
```
